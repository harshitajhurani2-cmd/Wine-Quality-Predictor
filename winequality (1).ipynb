```python
df['quality_label'] = df['quality'].apply(lambda x: 1 if x>=7 else 0)
```

```python
df
```

```python
df['quality_label'].value_counts()
```

```python
x = df.drop(['quality', 'quality_label'], axis=1)
y = df['quality_label']
```

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
```

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.20, random_state=42)
lr_model = LogisticRegression(max_iter = 1000)
```

```python
lr_model.fit(x_train, y_train)
```

```python
from sklearn.metrics import accuracy_score, f1_score, classification_report, recall_score, confusion_matrix, precision_score
```

```python
ypred = lr_model.predict(x_test)
```

```python
report = classification_report(y_test, ypred, output_dict=True)
report_df = pd.DataFrame(report).transpose()
print(report_df)
print("Accuracy=", accuracy_score(y_test, ypred))
print("Recall=", recall_score(y_test, ypred))
print("Precision=", precision_score(y_test, ypred))
print("f1_score=", f1_score(y_test, ypred))
print("Confusion Matrix = ", confusion_matrix(y_test, ypred))
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
scaler = StandardScaler()
x_train_scaled = scaler.fit_transform(x_train)
x_test_scaled = scaler.fit_transform(x_test)
```

```python
lr_model.fit(x_train_scaled, y_train)
```

```python
x_train_scaled, x_test_scaled, y_train, y_test = train_test_split(x, y, test_size=0.20, random_state=42)
```

```python
ypredscaled = lr_model.predict(x_test_scaled)
```

```python
classificationreport = classification_report(y_test, ypredscaled, output_dict= True)
```

```python
scaledclassify = pd.DataFrame(classificationreport).transpose()
```

```python
print(scaledclassify)
print("Accuracy=", accuracy_score(y_test, ypredscaled))
print("Recall=", recall_score(y_test, ypredscaled))
print("Precision=", precision_score(y_test, ypredscaled))
print("f1_score=", f1_score(y_test, ypredscaled))
print("Confusion Matrix = ", confusion_matrix(y_test, ypredscaled))
```

```python
from sklearn.neighbors import KNeighborsClassifier
from sklearn.tree import DecisionTreeClassifier
```

```python
x_train_scaled, x_test_scaled, y_train, y_test = train_test_split(x, y, test_size=0.20, random_state=42)
```

```python
knn = KNeighborsClassifier()
```

```python
knn.fit(x_train_scaled, y_train)
```

```python
ypred_knn = knn.predict(x_test_scaled)
```

```python
report = classification_report(y_test, ypred_knn, output_dict=True)
report_df = pd.DataFrame(report).transpose()
print(report_df)
print("Accuracy=", accuracy_score(y_test, ypred_knn))
print("Recall=", recall_score(y_test, ypred_knn))
print("Precision=", precision_score(y_test, ypred_knn))
print("f1_score=", f1_score(y_test, ypred_knn))
print("Confusion Matrix = ", confusion_matrix(y_test, ypred_knn))
```

```python
DTree = DecisionTreeClassifier()
```

```python
DTree.fit(x_train_scaled, y_train)
```

```python
ypred_dt = DTree.predict(x_test_scaled)
```

```python
report = classification_report(y_test, ypred_dt, output_dict=True)
report_df = pd.DataFrame(report).transpose()
print(report_df)
print("Accuracy=", accuracy_score(y_test, ypred_dt))
print("Recall=", recall_score(y_test, ypred_dt))
print("Precision=", precision_score(y_test, ypred_dt))
print("f1_score=", f1_score(y_test, ypred_dt))
print("Confusion Matrix = ", confusion_matrix(y_test, ypred_dt))
```

```python
# GridSearchCV
```

```python
from sklearn.model_selection import GridSearchCV
```

```python
classifier = GridSearchCV((DTree), {
    'criterion': ['gini', 'entropy', 'log_loss'],
    'splitter': ['best', 'random'], 
    'max_depth' : [3, 5, 7, 10, 12, None],
    'min_samples_split' : [2, 5, 10, 15], 
    'min_samples_leaf' : [1, 2, 4, 6, 8], 
    'class_weight' : [None, 'balanced']
}, cv=5, return_train_score= False
    
)
```

```python
classifier.fit(x_train_scaled, y_train)
```

```python
result = pd.DataFrame(classifier.cv_results_)
```

```python
result
```

```python
result[['param_criterion', 'param_splitter', 'param_max_depth',	'param_min_samples_leaf', 'param_min_samples_split','param_class_weight', 'mean_test_score' ]]
```

```python
print("Best Parameters :", classifier.best_params_)
print("Best Mean CV SCore", classifier.best_score_)
```

```python
best_dt = classifier.best_estimator_
```

```python
ypred_tuned = best_dt.predict(x_test_scaled)
```

```python
report = classification_report(y_test, ypred_tuned, output_dict=True)
report_df = pd.DataFrame(report).transpose()
print(report_df)
print("Accuracy=", accuracy_score(y_test, ypred_tuned))
print("Recall=", recall_score(y_test, ypred_tuned))
print("Precision=", precision_score(y_test, ypred_tuned))
print("f1_score=", f1_score(y_test, ypred_tuned))
print("Confusion Matrix = ", confusion_matrix(y_test, ypred_tuned))
```

```python
importance = pd.Series(best_dt.feature_importances_, index = x.columns).sort_values(ascending = False)
```

```python
print(importance)
```

```python
plt.figure(figsize=(10,6))
importance.plot(kind='bar', color= 'orange')
plt.ylabel("Importance Score")
plt.xlabel("Features")
plt.tight_layout()
plt.show()
```
